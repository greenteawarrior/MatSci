```python
dma1 = pd.read_csv('data/DMA/dma1.csv', header=[0,1]) 

dma1_fig, ax1 = plt.subplots(figsize=(10, 7))

ax1.plot(dma1['Time'], dma1['Program Temperature'], 'b.')
ax1.set_xlabel('Time (s)', fontsize=16)
ax1.set_ylabel('Program Temperature (C)', color='b', fontsize=16) # Make the y-axis label and tick labels match the line color.
for tl in ax1.get_yticklabels():
    tl.set_color('b')

ax2 = ax1.twinx()
ax2.plot(dma1['Time'], dma1['Tan Delta'], 'r.')
ax2.set_xlabel('Time (s)', fontsize=16)
ax2.set_ylabel('Tan Delta', color='r', fontsize=16) # Make the y-axis label and tick labels match the line color.
for tl in ax2.get_yticklabels():
    tl.set_color('r')
    
ax3 = ax1.twinx()
ax3.plot(dma1['Time'], dma1['Loss Modulus'], 'g.')
ax3.set_xlabel('Time (s)', fontsize=16)
ax3.set_ylabel('Loss Modulus', color='g', fontsize=16) # Make the y-axis label and tick labels match the line color.
ax3.spines['right'].set_position(('axes', 1.15))
ax3.get_yaxis().get_offset_text().set_position((1.25, 0))
for tl in ax3.get_yticklabels():
    tl.set_color('g')    

ax4 = ax1.twinx()
ax4.plot(dma1['Time'], dma1['Storage Modulus'], 'y.')
ax4.set_xlabel('Time (min)', fontsize=16)
ax4.set_ylabel('Storage Modulus', color='y', fontsize=16) # Make the y-axis label and tick labels match the line color.
ax4.spines['right'].set_position(('axes', 1.30))
ax4.get_yaxis().get_offset_text().set_position((1.40,0))
for tl in ax4.get_yticklabels():
    tl.set_color('y')    
    
axes = plt.gca()
# axes.set_xlim([300, 400])    
plt.rc('xtick', labelsize=16) 
plt.rc('ytick', labelsize=16)    

plt.title('DMA Trial 1', fontsize=18)
    
plt.show()
# dma1_fig.savefig("figures/dma1.pdf", bbox_inches='tight')
```

According to the DMA data above, we found that the piece was getting stiffer as temperature increased. This is not typical of materials unless there is a phase change involved. However this trial involved testing at higher temperatures and the sample seems to have gone through an irreversible phase change. We then conducted additional DMA trials to see if the previous phase change at the lower temperatures is reversible or not.
