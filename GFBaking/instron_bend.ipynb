### XG Bend Tests (Specimen 1, 2, 3)

Data taken on March 24, 2016

```python
# import
XG_bend1 = pd.read_csv('data/XG/XG_Bend_Data/XG_Bend_Specimen_RawData_1.csv', header=[0, 1])
XG_bend2 = pd.read_csv('data/XG/XG_Bend_Data/XG_Bend_Specimen_RawData_2.csv', header=[0, 1])
XG_bend3 = pd.read_csv('data/XG/XG_Bend_Data/XG_Bend_Specimen_RawData_3.csv', header=[0, 1])
```

```python
# plotting
pw = plt.figure(figsize=(10,7))
plt.scatter(XG_bend1['Extension'], XG_bend1['Load'], color='red')
plt.scatter(XG_bend2['Extension'], XG_bend2['Load'], color='orange')
plt.scatter(XG_bend3['Extension'], XG_bend3['Load'], color='yellow')

# formatting
plt.title('Extension vs. Load for Bend Test XG Bread Specimen 1, 2, 3', fontsize=18)
plt.xlabel('Extension (mm)', fontsize=18)
plt.ylabel('Load (N)', fontsize=18)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 
plt.legend(['Specimen 1', 'Specimen 2', 'Specimen 3'], loc=2, fontsize=16, bbox_to_anchor=(1.01, 1))
plt.show()

# uncomment if figure saving is desired
pw.savefig('figures/bend/XG_bend.pdf', format='pdf', bbox_inches='tight')
pw.savefig('figures/bend/XG_bend.png', format='png', bbox_inches='tight')
```

```python
def plot_bend(dataframes, expname='', save=False, filename='bend_plot'):
    
    # plotting
    pw = plt.figure(figsize=(10,7))
     
    colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'gray', 'black', 'magenta', 'cyan']
    for i in range(len(dataframes)):
        # offsets
        if expname == '8% Gluten (April 18)' and i == 0:
            print '1'
            plt.plot(dataframes[i]['Extension'].subtract(1.21266).multiply(-1), dataframes[i]['Load'], color=colors[i], alpha=1.0)
        elif expname == '8% Gluten (April 18)' and i == 4:
            print '2'
            plt.plot(dataframes[i]['Extension'].subtract(4.28932).multiply(-1), dataframes[i]['Load'], color=colors[i], alpha=1.0)
        else:
            plt.plot(dataframes[i]['Extension'].multiply(-1), dataframes[i]['Load'], color=colors[i], alpha=1.0)
        
    # formatting
    plt.title('%s Flexural Stress-Strain Response (4 Point Bend)' % (expname), fontsize=18)
    plt.xlabel('Extension (mm)', fontsize=18)
    plt.ylabel('Load (N)', fontsize=18)
    plt.rc('xtick', labelsize=15) 
    plt.rc('ytick', labelsize=15) 
    
    axes = plt.gca()
    axes.set_xlim([-1, 10])
    axes.set_ylim([-.0005, .001])
    
    plt.legend(['Specimen 1', 
                'Specimen 2', 
                'Specimen 3', 
                'Specimen 4', 
                'Specimen 5', 
                'Specimen 6', 
                'Specimen 7', 
                'Specimen 8'], 
                bbox_to_anchor=(1.01, 1), 
                loc=2, 
                fontsize=16)    
    plt.show()

    # saving the file
    if save:
        pw.savefig("figures/bend/%s.png" % filename, format='png', bbox_inches='tight')
        pw.savefig("figures/bend/%s.pdf" % filename, format='pdf', bbox_inches='tight')
    
    return
```
