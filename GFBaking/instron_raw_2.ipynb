### XG Compression Tests (Specimen 1, 2, 3, 4, 5, 6, 7, 8)

Data taken on March 21, 2016

```python
# import
xg_compress1 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen1.csv')
xg_compress2 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen2.csv')
xg_compress3 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen3.csv')
xg_compress4 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen4.csv')
xg_compress5 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen5.csv')
xg_compress6 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen6.csv')
xg_compress7 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen7.csv')
xg_compress8 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen8.csv')
```

### April 7th and April 14th. Binning

the plan:
* For each trial, Divide the data into reasonably sized bins by extension
    * For each bin, take the average of the loads
^ This can be done with .groupby and .agg
* Combine things from each trial
* Plot the binned curve

```python
g13_compress1.head()
```

```python
bins = [x/100.0 for x in range(-50, 1650, 50)]
print bins
```

```python
extension_bins = g13_compress1['Compressive extension mm'].groupby(pd.cut(g13_compress1['Compressive extension mm'], bins))
```

```python
aggregations = {
    'Compressive extension mm' : {
        'average_extension': 'mean',
        'num_extensions': 'count'
    },
    'Compressive load kgf': {
        'average_load': 'mean',
        'num_loads': 'count'
    }
}

agg13 = g13_compress1.groupby(pd.cut(g13_compress1['Compressive extension mm'], bins)).agg(aggregations)
# agg13
```
