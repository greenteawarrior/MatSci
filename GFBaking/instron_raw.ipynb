## Accessing and Plotting the Instron Data Captured in .raw Format

I did some work "by hand" opening the .raw files in a spreadsheet editor and saving the sample data as separate csv files.

(See other notebook for the situation with the .csv format)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import display
import numpy as np
import pprint as pp
```

### XG Compression Tests (Specimen 1, 2, 3, 4, 5, 6, 7, 8)

Data taken on March 21, 2016

```python
# import
compress1 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen1.csv')
compress2 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen2.csv')
compress3 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen3.csv')
compress4 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen4.csv')
compress5 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen5.csv')
compress6 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen6.csv')
compress7 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen7.csv')
compress8 = pd.read_csv('data/XG/XG_Compression_Data/XG_Compression_Specimen8.csv')
```

```python
# look at the header
compress8.head()
```

```python
# plotting
pw = plt.figure(figsize=(10,7))
plt.scatter(compress1['Extension mm'], compress1['Compressive load kgf'], color='red')
plt.scatter(compress2['Extension mm'], compress2['Compressive load kgf'], color='orange')
plt.scatter(compress3['Extension mm'], compress3['Compressive load kgf'], color='yellow')
plt.scatter(compress4['Extension mm'], compress4['Compressive load kgf'], color='green')
plt.scatter(compress5['Extension mm'], compress5['Compressive load kgf'], color='blue')
plt.scatter(compress6['Extension mm'], compress6['Compressive load kgf'], color='purple')
plt.scatter(compress7['Extension mm'], compress7['Compressive load kgf'], color='gray')
plt.scatter(compress8['Extension mm'], compress8['Compressive load kgf'], color='black')

# formatting
plt.title('Extension vs. Load for Compression Test XG Bread Specimen 1, 2, 3, 4, 5, 6, 7, 8', fontsize=18)
plt.xlabel('Extension (mm)', fontsize=18)
plt.ylabel('Load (N)', fontsize=18)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 
plt.legend(['Specimen 1', 'Specimen 2', 'Specimen 3', 'Specimen 4', 'Specimen 5', 'Specimen 6', 'Specimen 7', 'Specimen 8'], loc=1, fontsize=16)
plt.show()

# uncomment if figure saving is desired
# pw.savefig('figures/XG_Compression_Sp12345678.pdf', format='pdf')
# pw.savefig('figures/XG_Compression_Sp12345678.png', format='png')
```
